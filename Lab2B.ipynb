```python
def mergeSort(arr, depth=1):
    if len(arr) > 1:
        mid = len(arr) // 2  # Find the middle index
        left = arr[:mid]      # Divide the array into two halves
        right = arr[mid:]

        mergeSort(left, depth + 1)  # Recursively sort the left half
        mergeSort(right, depth + 1) # Recursively sort the right half

        i = j = k = 0  # Indices for left, right, and merged array

        # Merge the two sorted halves
        while i < len(left) and j < len(right):
            if left[i] < right[j]:  
                arr[k] = left[i]  # Copy smaller element from left half
                i += 1
            else:
                arr[k] = right[j] # Copy smaller element from right half
                j += 1
            k += 1

        # Copy any remaining elements from the left half
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1

        # Copy any remaining elements from the right half
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1
        
        # Print the state of the list at each merge step
        print(f"Pass {depth}: {arr}")

# Main execution
arr = [3, 5, 2, 9, 8, 1, 6, 4, 7]
print(f"Original List: {arr}")
mergeSort(arr)
print(f"Sorted List: {arr}")
```

```python
def mergesort(arr, depth=0):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = mergesort(arr[:mid], depth + 1)
    right = mergesort(arr[mid:], depth + 1)

    merged = merge(left, right, depth) 
    return merged

def merge(left, right, depth):
    merged = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    merged.extend(left[i:])
    merged.extend(right[j:])
    if depth > 0:
        print(f"Pass {depth}: {merged}")
    return merged

arr = [3, 5, 2, 9, 8, 1, 6, 4, 7]
print(f"Original List: {arr}")

sortedarr = mergesort(arr)
print(f"Sorted List: {sortedarr}")
```

```python
arr = [23, 17, 21, 3, 42, 9, 13, 1, 2, 7, 35, 4]
print(f"Original List: {arr}")

def quicksort(arr, depth=0): 
    if len(arr) <= 1:
        return arr
    
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    
    sorted_arr = quicksort(left, depth + 1) + middle + quicksort(right, depth + 1)
    print(f"Pass {depth}: {sorted_arr}")
    return sorted_arr

sortedarr = quicksort(arr)
print(f"Sorted List: {sortedarr}")
```

```python
arr = [23, 17, 21, 3, 42, 9, 13, 1, 2, 7, 35, 4]
print(f"Original List: {arr}")

def heapify(arr, n, i, depth):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left
    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest, depth + 1)

def heapsort(arr):
    n = len(arr)
    
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i, 1)
    print(f"Max Heap: {arr}")
    
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0, 1)
        print(f"Pass {n - i}: {arr}")

heapsort(arr)
print(f"Sorted List: {arr}")
```

```python
def heapsort(arr):
    print(f"Original List: {arr}")
    
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        largest = i
        while True:
            left = 2 * largest + 1
            right = 2 * largest + 2
            swap = largest
            if left < n and arr[left] > arr[swap]:
                swap = left
            if right < n and arr[right] > arr[swap]:
                swap = right
            if swap == largest:
                break
            arr[largest], arr[swap] = arr[swap], arr[largest]
            largest = swap
    
    print(f"Max Heap: {arr}")
    
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        size = i
        largest = 0
        while True:
            left = 2 * largest + 1
            right = 2 * largest + 2
            swap = largest
            if left < size and arr[left] > arr[swap]:
                swap = left
            if right < size and arr[right] > arr[swap]:
                swap = right
            if swap == largest:
                break
            arr[largest], arr[swap] = arr[swap], arr[largest]
            largest = swap
        print(f"Pass {n - i}: {arr}")
    
    print(f"Sorted List: {arr}")

arr = [23, 17, 21, 3, 42, 9, 13, 1, 2, 7, 35, 4]
heapsort(arr)
```

```python
def mult(a, b):
    if a < 10 or b < 10:
        return a * b

    n = max(len(str(a)), len(str(b)))
    m = n // 2

    a1 = a // 10**m
    a2 = a % 10**m
    b1 = b // 10**m
    b2 = b % 10**m

    ac = mult(a1, b1)
    bd = mult(a2, b2)
    ad_bc = mult(a1 + a2, b1 + b2) - ac - bd

    return ac * 10**(2 * m) + ad_bc * 10**m + bd

a = 278
b = 6543

print(mult(a, b))
```

```python
def mult(a, b, depth=0):
    print(f"Step {depth}: Multiplying {a} and {b}")
    
    if a < 10 or b < 10:
        print(f"Step {depth}: Base case {a} * {b} = {a * b}")
        return a * b

    n = max(len(str(a)), len(str(b)))
    m = n // 2

    a1 = a // 10**m
    a2 = a % 10**m
    b1 = b // 10**m
    b2 = b % 10**m

    print(f"Step {depth}: Splitting {a} -> {a1}, {a2} and {b} -> {b1}, {b2}")

    ac = mult(a1, b1, depth + 1)
    bd = mult(a2, b2, depth + 1)
    ad_bc = mult(a1 + a2, b1 + b2, depth + 1) - ac - bd

    result = ac * 10**(2 * m) + ad_bc * 10**m + bd
    print(f"Step {depth}: Result {result}")
    
    return result

a = 679889
b = 234675

print(f"Final Product: {mult(a, b)}")
```
