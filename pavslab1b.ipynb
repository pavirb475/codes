```python
#1D ARRAY
arr=[2,3,4,5,6]

#Accessing an element by index.
element=arr[3]
print("The element at index 3 is",element)

#Inserting an element in the beginning, middle and end
def insertarr(arr,num,index):
    arr.append(0) 
    for i in range(len(arr) - 2, index - 1, -1):
        arr[i + 1] = arr[i]
    arr[index]= num
    print(arr)
    
print("Inserting elements:")
insertarr(arr, 9, 0) #beginning
insertarr(arr, 10, 3) #middle
insertarr(arr, 11, 7) #end

#Transposing an element
def transposearr(arr,ind1,ind2):
    x=arr[ind1]
    y=arr[ind2]
    arr[ind1]=y
    arr[ind2]=x
    print(arr)
    
print("Transposing elements of index 0 and 5:")
transposearr(arr, 0, 5)

#Deleting an element
def deletearr(arr,index):
    for i in range(index, len(arr) - 1):
        arr[i] = arr[i+1]
    arr.pop()
    print(arr)
print("Deleting element of index 4:")
deletearr(arr,4)
```

```python
#2D ARRAY

arr = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

#Accessing an element by index
element = arr[2][1]
print("The element at index (2,1) is", element)

#Transposing an element
def transposearr(arr, row, col):
    arr[row][col], arr[col][row] = arr[col][row], arr[row][col]
    print(arr)

print("Transposing elements (2,1) with (1,2):")
transposearr(arr, 2, 1)
```

```python
n = int(input("Enter a number: ")) 
sq = n*n 
num = n  
d = len(str(n))  
print("Number of digits:", d)

lasidigi = sq%(10**len(str(n)))
if lasidigi!= num: 
    print(f"{n} is NOT an automorphic number")  
else:  
    print(f"{n} is an automorphic number")  
```

```python
arr1 = [
    [1, 2, 3],
    [4, 1, 6],
    [7, 8, 1]
]
arr2 = [
    [1, 2, 3],
    [4, 3, 6],
    [7, 8, 1]
]

def toeplitzcheck(arr):
    val=arr[0][0]
    for i in range(1, len(arr)):
        if arr[i][i]!= val:
            print("Not a Toeplitz Matrix")
            return
    print("It is a Toeplitz Matrix")
    
print("Matrix 1:",arr1)
toeplitzcheck(arr1)
print("Matrix 2:",arr2)
toeplitzcheck(arr2)
```

```python
arr1=[2,2,3,4,5,6]
arr2=[2,7,4,9,6]

def checksorted(arr):
    for i in range(0,len(arr)-1):
        if arr[i]>arr[i+1]:
            print("Not sorted")
            return
    print("Sorted")   
    
print("Array 1:",arr1)
checksorted(arr1)
print("Array 2:",arr2)
checksorted(arr2)
```

```python
arr1 = [2, 2, 3, 4, 5, 6]
arr2 = [2, 7, 4, 9, 6]

def checksorted(arr, ind=0):
    if ind >= len(arr) - 1:  #base case
        print("SORTED")
        return
    if arr[ind] > arr[ind + 1]: 
        print("NOT SORTED")
        return
    checksorted(arr, ind + 1)  

print("Array 1:", arr1)
checksorted(arr1)
print("Array 2:", arr2)
checksorted(arr2)
```
